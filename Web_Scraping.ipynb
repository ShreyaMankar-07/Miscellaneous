```python
final_df = pd.DataFrame()
for j in range(1,300):
    url = requests.get(f"https://www.ambitionbox.com/list-of-companies?page={j}", headers=headers).text
    soup = BeautifulSoup(url, 'lxml')
    
    #for info present in primaryInformation div
    company = soup.find_all('div',class_="companyCardWrapper__primaryInformation")
    
    name = []
    rating = []
    type = []
    hq = []
    old = []
    emp_count = []
    
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)
        try:
            rating.append(i.find('span',class_="companyCardWrapper__companyRatingValue")) #need more extraction
        except:
            rating.append(np.nan)
        try:
            type.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[-3].strip())
        except:
            type.append(np.nan)
        try:
            hq.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[-1].strip())
        except:
            hq.append(np.nan)
        try: 
            old.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[-2].strip())
        except:
            old.append(np.nan)
        try:
            emp_count.append(i.find_all('span',class_='companyCardWrapper__interLinking')[0].text.strip().split('|')[1].strip())
        except:
            emp_count.append(np.nan)
    
    #to extract individual rating from 'rating' list
    rate_count = []
    for i in range(len(rating)):
        rate_count.append(rating[i].text.strip())
        
    #review info present in tertiaryInformation div
    other_info = soup.find_all('div',class_="companyCardWrapper__tertiaryInformation")
    
    reviews=[]
    for i in other_info:
        reviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[0].text.strip())
        
    dict = {"company_name":name,"rating":rate_count,"reviews":reviews,"company_type":type,'headquarters':hq,'years_old':old,"emp_count":emp_count}
    
    df = pd.DataFrame(dict)
    final_df = pd.concat([final_df,df], ignore_index=True)

final_df
```
